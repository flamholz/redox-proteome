```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Harris and Theriot 2018 estimate the SA/V ratio of a cell as a function of growth rate.  
lambdas = np.linspace(0.05, 2, 100)  # hr^-1 units 
SA_V = 9.3 -2.8*lambdas              # um^-1 

# Assuming the membrane thickness is very small compared to cell width,
# that the two leaflets are the same thickness, and that they differ
# negligibly in surface area. 
mem_thickness = 0.005 # um
mem_v_fraction_1leaflet = SA_V*mem_thickness
mem_v_fraction_2leaflet = 2*mem_v_fraction_1leaflet


# lipid fraction is less dense than cytoplasm
rho_cyto = 1.1 # g/mL
rho_mem = 1.0  # g/mL
mem_m_fraction_1leaflet = mem_v_fraction_1leaflet*rho_mem/rho_cyto
mem_m_fraction_2leaflet = mem_v_fraction_2leaflet*rho_mem/rho_cyto
```

```python
plt.figure()
plt.plot(lambdas, 100*mem_v_fraction_1leaflet, label='1 leaflet')
plt.plot(lambdas, 100*mem_v_fraction_2leaflet, label='2 leaflet')
plt.xlabel('Growth rate (hr$^{-1}$)')
plt.ylabel('Membrane volume fraction (%)')
plt.show()
```
